# Auswertung für großes Datenset: messdatenZeit-Confidence1.1
Wurde gegen 12:00 Uhr aufgenommen

```python
import pandas as pd
import numpy as np
```

Datenset lesen

```python
df2 =pd.read_csv("messdatenZeit-Confidence1.1.csv")
df2 = df2.rename(columns={'confidence_YOLO': 'Schwellwert'})
df2.head()
```

gemessene Dimensonen hinzufügen

```python
referenzwerte = {
    'Würfel': {'max': 2.3, 'min': 2.3},
    'Ball': {'max': 5.968, 'min': 5.968},
    'Metallplatte': {'max': 19.0, 'min': 11.0},
    'Schraubenzieher': {'max': 9.5, 'min': 3.0},
    'Schraubenzieher Griff': {'max': 9.5, 'min': 3.0},
    'Schaumstoff Ding': {'max': 38.0, 'min': 4.5}
}
```

```python
df2['maxSide'] = df2['Objekt'].map(lambda x: referenzwerte.get(x, {}).get('max', None))
df2['minSide'] = df2['Objekt'].map(lambda x: referenzwerte.get(x, {}).get('min', None))
df2.head()
```

```python
df2.info()
```

Zeit bis Objekt als stabil erkannt wurde

```python
big_data_mittelwertZeit = df2.groupby('Schwellwert')['ElapsedTime [s]'].agg(
    Dauer_Durchschnitt='mean',
    Standardabweichung='std',
    Minimum='min', 
    Maximum='max'
).reset_index()
big_data_mittelwertZeit = big_data_mittelwertZeit.sort_values(
    by='Dauer_Durchschnitt',
    ascending=True
)
big_data_mittelwertZeit
```

Dimensons-Abweichungen berechnen

```python
df2['Reale_Diagonale [cm]'] = np.sqrt(df2['maxSide']**2 + df2['minSide']**2)
df2['Aweichung maxSide[cm]'] = np.sqrt((df2['maxSide'] - df2['xSize [cm]'])**2 )
df2['Aweichung minSide[cm]'] = np.sqrt( (df2['minSide'] - df2['ySize [cm]'])**2)
df2['Abweichung [cm]'] = np.sqrt((df2['maxSide'] - df2['xSize [cm]'])**2 + (df2['minSide'] - df2['ySize [cm]'])**2)
df2['Abweichung [%]'] = (df2['Abweichung [cm]'] / df2['Reale_Diagonale [cm]']) * 100

df2.head()
```

Nach Aweichung sorrtieren

```python
# Gruppieren nach Schwellwert
data_mittelwertAbweichung = df2.groupby('Schwellwert')['Abweichung [%]'].agg(
    Abweichung_Durchschnitt='mean',
    Standardabweichung='std',
)

# Nach durchschnittlicher Abweichung sortieren
data_mittelwertAbweichung = data_mittelwertAbweichung.sort_values(
    by='Abweichung_Durchschnitt',
    ascending=True
).reset_index()
data_mittelwertAbweichung
```

# Kleines Datenset ~9:00 Uhr mit großem Datenset ~12:00 Uhr vergleichen

Objekte die nicht in großem Datenset sind rausziehen

```python
label = pd.read_csv('Labels.csv')
label.head(12)
messreihe = pd.read_csv('messdatenZeit-Confidence.csv')

df1 = pd.concat([label, messreihe], axis=1)
df1 =  df1.replace(5968.0, 5.968)
#data.to_csv('messdatenConfidence.csv', index= False)




df_klein = df1.copy()
df_klein = df_klein[df_klein['Objekt'] != 'batterie']
df_klein['Objekt'] = df_klein['Objekt'].replace('Schraubenzieher Griff', 'Schraubenzieher')

print(df_klein.groupby('Objekt')['ElapsedTime [s]'].agg(
    DurchschnitZeit = 'mean',
    stdTime = 'std'
).reset_index())
df_klein = df_klein.groupby('confidence_YOLO')['ElapsedTime [s]'].agg(
    DurchschnitZeit = 'mean',
    stdTime = 'std'
).reset_index()
df_klein
```

Große Liste komprimieren, sodass nur erster Wert von Objekt/Confidnce in df ist

```python
# c = []
# i = 2
# for n in range(0,60):
#     #print(i)
#     c.append(i)
#     i += 10

c = list(range(2, 500, 10))

df_komprimiert = df2.iloc[c].reset_index(drop=True)
df_komprimiert = df_komprimiert.groupby('Schwellwert')['ElapsedTime [s]'].agg(
    Durchschnitt_s = 'mean',
    std = 'std'
).reset_index()
df_komprimiert
```

Zusammenbringer beider Dfs

```python
df_vergleich = pd.merge(
    df_komprimiert,         # 🔹 Erster DataFrame
    df_klein,                # 🔹 Zweiter DataFrame
    left_on="Schwellwert",  # 🔹 Spalte im ersten DF, die als Verknüpfung verwendet wird
    right_on="confidence_YOLO",  # 🔹 Entsprechende Spalte im zweiten DF
    how="inner"             # 🔹 Verknüpfungsart: nur Übereinstimmungen (wie INNER JOIN in SQL)
)
df_vergleich = df_vergleich.drop(columns=['confidence_YOLO', 'std', 'stdTime'])
df_vergleich = df_vergleich.rename(columns={'Durchschnitt_s': 'Mean[s] 12:00', 'DurchschnitZeit': 'Mean[s] 9:00'})
df_vergleich = df_vergleich[["Schwellwert", "Mean[s] 9:00", "Mean[s] 12:00"]]

df_vergleich
```

```python
df_vergleich['Differenz [s]'] = (df_vergleich['Mean[s] 9:00']  - df_vergleich['Mean[s] 12:00']) 
df_vergleich
```
